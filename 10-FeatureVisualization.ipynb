```python
import os
import time
import pandas as pd
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# load attributes 
data_dir = os.getcwd() + '\\Data\\'
train_data = pd.read_csv(data_dir + 'train.csv')
```

```python
# reading other features
temp = train_data.copy().sort_values(by = ['species']).copy()
margins = temp.values[:,2:66].copy()
shapes = temp.values[:,66:130].copy()
textures =temp.values[:,130:199].copy()
```

```python
vis_margins = np.array(np.reshape(margins, (temp.shape[0], 8, 8)), dtype=float)
count = 64
fig = plt.figure(figsize=(10, 10))
for i in range(count):
    sub = plt.subplot(8, 8, i + 1)
    sub.set_xticks(())
    sub.set_yticks(())
    sub.set_title(temp.values[i,1], size=6)
    sub.imshow(vis_margins[i], cmap=plt.get_cmap('gray_r'))
    
fig.tight_layout(pad = 0.1)
plt.show()
```

```python
vis_shapes = np.array(np.reshape(shapes, (temp.shape[0], 8, 8)), dtype=float)
count = 64
fig = plt.figure(figsize=(10, 10))
for i in range(count):
    sub = plt.subplot(8, 8, i + 1)
    sub.set_xticks(())
    sub.set_yticks(())
    sub.set_title(temp.values[i,1], size=6)
    sub.imshow(vis_shapes[i], cmap=plt.get_cmap('gray_r'))
    
fig.tight_layout(pad = 0.1)
plt.show()
```

```python
vis_textures = np.array(np.reshape(textures, (temp.shape[0], 8, 8)), dtype=float)
count = 64
fig = plt.figure(figsize=(10, 10))
for i in range(count):
    sub = plt.subplot(8, 8, i + 1)
    sub.set_xticks(())
    sub.set_yticks(())
    sub.set_title(temp.values[i,1], size=6)
    sub.imshow(vis_textures[i], cmap=plt.get_cmap('gray_r'))
    
fig.tight_layout(pad = 0.1)
plt.show()
```
